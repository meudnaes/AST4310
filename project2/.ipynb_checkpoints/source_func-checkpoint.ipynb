```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from astropy import units
from scipy.integrate import cumtrapz, trapz
from astropy.visualization import quantity_support
from IPython.display import set_matplotlib_formats
quantity_support()
set_matplotlib_formats('svg')
```

```python
freq = np.linspace(-5,5,101)
extinction = (np.exp(-freq**2)+ 0.5*freq/50 + 0.1)/units.m/100

depth = np.logspace(0, 4, 200)*units.km
source_function = depth.value/100
#source_function = (np.linspace(0,100,len(depth)))*5
#source_function = np.exp(-depth.value/500)

#Make extinction the same at all depths (unrealistic)
extinction = np.tile(extinction, (len(depth), 1))

height = -depth + depth.max()

tau = cumtrapz(extinction, depth, axis=0, initial = 1e-3)
Intensity = trapz(source_function[:,np.newaxis]*np.exp(-tau), tau, axis=0)

#Plots
fig, ax = plt.subplots(2,2, figsize=(9,6))
ax[0,0].plot(freq, extinction[0])
ax[0,0].set_title('Extinction')

ax[1,1].plot(height, source_function)
ax[1,1].set_title('Source function')

ax[0,1].plot(height, tau[:,0])
ax[0,1].plot(height, tau[:,50], 'r')
ax[0,1].set_title(r'Optical depth $\tau$')

ax[1,0].plot(freq, Intensity)
ax[1,0].set_title('Intensity')
plt.tight_layout()
```
