```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from astropy import units
from astropy import constants
from astropy.visualization import quantity_support
from IPython.display import set_matplotlib_formats
from astropy.modeling.blackbody import BlackBody

quantity_support()
set_matplotlib_formats('svg')
plt.rc('legend', frameon=False)
```

```python
temp = units.Quantity([5000,8000], unit='K')
bb = BlackBody(temp)

i_units_nu = 'W m-2 Hz-1 sr-1'
bb(500 * units.nm).to(i_units_nu)

i_units = 'W m-2 nm-1 sr-1'
bb_lambda = BlackBody(temp, scale=1. * units.Unit(i_units))
bb_lambda(500 * units.nm)

wave = numpy.arange(100,2000,100) * units.nm

bb_lambda(wave[:, numpy.newaxis])

from cycler import cycler
from matplotlib import cm #colormap

radiation = BlackBody(temp, scale=1.*units.Unit(i_units))(wave[:,numpy.newaxis])
fig, ax = plt.subplots()
custom_cycler = cycler('color', cm.plasma(numpy.linspace(0,1,len(temp)))) #going to go through the whole cm
ax.set_prop_cycle(custom_cycler) #change default colours of plots
ax.plot(wave, radiation.to('kW m-2 nm-1 sr-1'));
```
